```python

stats = df_result.groupby(['dataset_name','classifier_name','hyperparameter'])['time'].agg(['mean', 'count', 'std'])

ci95_hi = []
ci95_lo = []
ci95_interval = []

for i in stats.index:
    m, c, s = stats.loc[i]
    
    #ci95_hi.append(m + 1.96*s/math.sqrt(c))
    #ci95_lo.append(m - 1.96*s/math.sqrt(c))
    ci95_interval.append(1.96*s/math.sqrt(c))

#stats['ci95_hi'] = ci95_hi
#stats['ci95_lo'] = ci95_lo
stats['ci95_interval'] = ci95_interval
#stats['ci95_interval2'] = ci95_interval
stats["hour"]=np.floor(stats["mean"]/(60*60)).astype(int)
stats["min"]=np.floor(stats["mean"]/60-stats["hour"]*60).astype(int)
stats["seg"]=np.floor(stats["mean"]-stats["hour"]*(60*60)-stats["min"]*60).astype(int)
stats["ms"]=np.floor((stats["mean"]*1000)-stats["hour"]*(60*60*1000)-stats["min"]*(60*1000)-(stats["seg"]/1000)).astype(int)
stats["time_rf"]=""+stats["hour"].astype(str)+"h "+stats["min"].astype(str)+"m "+stats["seg"].astype(str)+"s "+stats["ms"].astype(str)+"ms"

stats=pd.DataFrame(stats[["time_rf","mean","ci95_interval"]]).round(2)

stats
```
